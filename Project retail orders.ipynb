```python
import pandas as pd 
df = pd.read_csv('/Users/ikawahyuni/Downloads/ectracted_orders/orders.csv')
df.head(20)
```

```python
#in pandas only NaN will be ackowledge, so 'unknown' and 'Not Available' need to replace 
#first we want to know how many unique data din 'Ship Mode' 
df['Ship Mode'].unique()
```

```python
#to replace 'Not Available' and 'Unkown', the simple one is using this code below \
#na_values =['Not Available', 'unknown' will be automatically read into NaN
df = pd.read_csv('/Users/ikawahyuni/Downloads/ectracted_orders/orders.csv', na_values =['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#check if Not Available and unknown values already replace to NaN 
df.head(20)
```

```python
#change name of columns to lower case 
#and convert space to '_'
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
#merged code = df.columns = df.columns.str.lower().str.replace(' ','_')
df.columns
```

```python
#Count discount
df['discount']=df['list_price']*df['discount_percent']*0.01
df.head()#to check if the discount column is in the df
```

```python
#Count Sale_Price
df['sale_price'] = df['list_price']-df['discount']


#Count the profit of the company 
# Company brought the product with price in 'cost_price' and sell it with price in the 'sale_price' 
#so to count profit we calculate 'sale_price' - 'cost_price' 
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
#convert order_date type 'object' to datetime
df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.dtypes
```

```python
#drop unnessary columns 
df.drop(['list_price', 'cost_price', 'discount_percent'], axis = 1, inplace = True)
df
```

```python
from sqlalchemy import create_engine, text
import pandas as pd

# Gunakan kredensial yang SAMA dengan di DBeaver
server = 'localhost'
port = '5433'  # port yang sama dengan DBeaver
database = 'postgres'
username = 'postgres'
password = 'ikawahyuni1234'

# Connection string PostgreSQL
connection_string = f"postgresql://{username}:{password}@{server}:{port}/{database}"

engine = create_engine(connection_string)

# Tes koneksi
try:
    with engine.connect() as conn:
        result = conn.execute(text("SELECT version()")).fetchone()
        print("âœ… Koneksi berhasil!")
        print(result[0])
except Exception as e:
    print("âŒ Koneksi gagal:")
    print(e)
```

```python
try:
    df.to_sql(
        name='orders',           # nama tabel di PostgreSQL
        con=engine,              # koneksi engine
        if_exists='replace',     # 'replace', 'append', atau 'fail'
        index=False              # tidak menyimpan index sebagai kolom
    )
    print("âœ… Data berhasil ditransfer ke PostgreSQL!")
    print(f"Jumlah baris: {len(df)}")
except Exception as e:
    print("âŒ Transfer gagal:")
    print(e)
```
